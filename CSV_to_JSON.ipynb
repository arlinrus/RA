```python
import json
import csv
from google.colab import drive
drive.mount('/content/drive')
```

```python
def csv_covert(csv_path, json_path):
    csv_data = []
    with open(csv_path, 'r', encoding='utf-8') as csv_file:
        csv_reader = csv.DictReader(csv_file)
        for row in csv_reader:
            csv_data.append(row)

    with open(json_path, 'w', encoding='utf-8') as json_file:
        json.dump(csv_data, json_file, indent=4)

csv_file_path = '/content/drive/MyDrive/Colab Notebooks/adult_data_train.csv'


json_path ='/content/drive/MyDrive/Colab Notebooks/adult_dataJSON_train.json'
csv_covert(csv_file_path, json_path)

print("Ready", json_path)
```
